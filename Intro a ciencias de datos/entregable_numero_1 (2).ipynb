# **3. Formulación de preguntas**

¿Qué equipo es el que más partidos ha ganado en la historia de los mundiales de la FIFA?

¿Cuál es el partido en el que más goles se marcaron?

¿Cuál es el partido con mayor diferencia de gol?

```python
# 1. ¿Qué equipo es el que más partidos ha ganado en la historia de los mundiales de la FIFA?

# Determine the winner for each match in the cleaned dataframe
def get_winner(row):
    if row['Home Team Goals'] > row['Away Team Goals']:
        return row['Home Team Name']
    elif row['Away Team Goals'] > row['Home Team Goals']:
        return row['Away Team Name']
    else:
        return 'Draw'

df_cleaned['Winner'] = df_cleaned.apply(get_winner, axis=1)

# Combine 'Germany FR' and 'Germany'
df_cleaned['Winner'] = df_cleaned['Winner'].replace('Germany FR', 'Germany')


# Count the number of wins for each team, excluding draws
team_wins = df_cleaned[df_cleaned['Winner'] != 'Draw']['Winner'].value_counts()

print("Equipo con más partidos ganados:")
display(team_wins.head(1))
```

```python
# 2. ¿Cuál es el partido en el que más goles se marcaron?

# Calculate the total goals per match
df_cleaned['Total Goals'] = df_cleaned['Home Team Goals'] + df_cleaned['Away Team Goals']

# Find the match with the maximum total goals
match_with_most_goals = df_cleaned.loc[df_cleaned['Total Goals'].idxmax()]

print("\nPartido con más goles marcados:")
display(match_with_most_goals[['Year', 'Stage', 'Home Team Name', 'Away Team Name', 'Total Goals']])
```

```python
# 3. ¿Cuál es el partido con mayor diferencia de gol?

# Calculate the goal difference per match
df_cleaned['Goal Difference'] = abs(df_cleaned['Home Team Goals'] - df_cleaned['Away Team Goals'])

# Find the match with the maximum goal difference
match_with_largest_goal_difference = df_cleaned.loc[df_cleaned['Goal Difference'].idxmax()]

print("\nPartido con mayor diferencia de gol:")
display(match_with_largest_goal_difference[['Year', 'Stage', 'Home Team Name', 'Away Team Name', 'Goal Difference']])
```

# **4. Visualización de gráficos**

```python
# Get the top 5 teams with most wins
top_6_teams = team_wins.head(6)

# Create a bar chart using Plotly Express
import plotly.express as px
fig = px.bar(top_6_teams,
             x=top_6_teams.index,
             y=top_6_teams.values,
             labels={'x': 'Equipo', 'y': 'Número de Partidos Ganados'},
             title='Top 6 Equipos con Más Partidos Ganados en Mundiales')

# Display the chart
fig.show()
```

**Gráfico de los 6 Equipos con Más Partidos Ganados**

Este gráfico de barras muestra los 6 equipos nacionales que han ganado la mayor cantidad de partidos en la historia de la Copa Mundial de la FIFA, según los datos analizados. El eje horizontal representa el nombre del equipo, mientras que el eje vertical indica el número total de partidos ganados por cada equipo.

Se puede observar claramente qué equipos han tenido un rendimiento histórico más destacado en términos de victorias en el torneo. Brasil lidera la lista, seguido por otros equipos con una rica historia en los mundiales. Este tipo de visualización permite comparar rápidamente el éxito de los equipos más laureados a lo largo de las ediciones de la Copa del Mundo.
