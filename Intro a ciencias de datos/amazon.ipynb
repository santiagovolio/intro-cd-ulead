Crea un gráfico de línea interactivo utilizando Plotly para visualizar la tendencia de "Total Sales" a lo largo del tiempo, utilizando la columna "Date" como eje x. Asegúrate de que la columna "Date" esté en formato datetime y filtra los datos para incluir solo el año 2025.

## Preparación de datos

### Subtask:
Asegurarse de que la columna "Date" esté en formato datetime y filtrar los datos para incluir solo el año 2025.


**Reasoning**:
The subtask requires converting the 'Date' column to datetime format and then filtering the DataFrame to include only data from the year 2025. These steps can be combined into a single code block.



```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')
df_2025 = df[df['Date'].dt.year == 2025]
display(df_2025)
```

## Crear el gráfico

### Subtask:
Generar un gráfico de línea utilizando Plotly para visualizar la tendencia de "Total Sales" a lo largo del tiempo.


**Reasoning**:
Generate a line plot using Plotly to visualize the trend of "Total Sales" over time.



```python
import plotly.express as px

df_2025['Week'] = df_2025['Date'].dt.to_period('W').dt.start_time
weekly_sales = df_2025.groupby('Week')['Total Sales'].sum().reset_index()

fig = px.line(weekly_sales, x='Week', y='Total Sales', title='Total Sales Trend in 2025 (Weekly)', color_discrete_sequence=['red'])
fig.show()
```
